```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv

img = cv.imread('sst.jpg', 0)
rows, cols = img.shape

#Plot the original image
plt.subplot(1, 2, 1)
plt.title("Original")
plt.imshow(img)

M = np.float32([[1, 0, 100],[0, 1, 50]])
transl = cv.warpAffine(img, M, (cols, rows))

#Plot the translated image
plt.subplot(1, 2, 2)
plt.title("Translated Image")
plt.imshow(transl)

cv.waitKey(0)
cv.destroyAllWindows()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv

img = cv.imread("sst.jpg", 0)

# Plot the original image
plt.subplot(1, 2, 1)
plt.title("Original")
plt.imshow(img, cmap="gray")

rows, cols = img.shape
M = np.float32([[1, 0, 0], [0, -1, rows], [0, 0, 1]])
reflected_img = cv.warpPerspective(img, M, (int(cols), int(rows)))

# Plot the reflected image
plt.subplot(1, 2, 2)
plt.title("Reflected Image")
plt.imshow(reflected_img, cmap="gray")

cv.imwrite("reflection_out.jpg", reflected_img)

plt.show()
cv.destroyAllWindows()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv

img = cv.imread("sst.jpg", 0)
rows, cols = img.shape

# Plot the original image
plt.subplot(1, 2, 1)
plt.title("Original")
plt.imshow(img, cmap="gray")

img_rotation = cv.warpAffine(
    img,
    cv.getRotationMatrix2D((cols / 2, rows / 2), 30, 0.6),
    (cols, rows),
)

# Plot the rotated image
plt.subplot(1, 2, 2)
plt.title("Rotated Image")
plt.imshow(img_rotation, cmap="gray")

cv.imwrite("rotation_out.jpg", img_rotation)

plt.show()
cv.destroyAllWindows()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv

img = cv.imread("sst.jpg", 0)
rows, cols = img.shape

plt.subplot(1, 2, 1)
plt.title("Original")
plt.imshow(img, cmap="gray")

M = np.float32([[1, 0.5, 0], [0, 1, 0], [0, 0, 1]])
sheared_img = cv.warpPerspective(img, M, (int(cols * 1.5), int(rows * 1.5)))

plt.subplot(1, 2, 2)
plt.title("Sheared Image")
plt.imshow(sheared_img, cmap="gray")

cv.imshow("img", sheared_img)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
# Importing libraries
import cv2

image = cv2.imread('sst.jpg')

cv2.imshow('Original Image', image)
cv2.waitKey(0)

# Gaussian Blur
Gaussian = cv2.GaussianBlur(image, (7, 7), 0)
cv2.imshow('Gaussian Blurring', Gaussian)
cv2.waitKey(0)

# Median Blur
median = cv2.medianBlur(image, 5)
cv2.imshow('Median Blurring', median)
cv2.waitKey(0)

# Bilateral Blur
bilateral = cv2.bilateralFilter(image, 9, 75, 75)
cv2.imshow('Bilateral Blurring', bilateral)
cv2.waitKey(0)

cv2.destroyAllWindows()
```
