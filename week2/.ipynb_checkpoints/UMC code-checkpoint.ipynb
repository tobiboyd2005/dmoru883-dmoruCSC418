```python
import cv2
import matplotlib.pyplot as plt

# Dictionary storing user data
users = {
    "Enase": {"surname": "Okonedo", "image": "enase.jpg"},
    "Darlington": {"surname": "Argholor", "image": "darlington.jpg"},
    "Bursar": {"surname": "Williams", "image": "bursar.png"},
}

# User input
username = input("Enter your first name: ").strip()
password_input = input("Enter your password (sum of surname letters): ").strip()

# Check if the user exists
if username in users:
    correct_password = str(len(users[username]["surname"]))  # Compute correct password
    
    if password_input == correct_password:
        print("Login successful!")

        # Load the image
        image_path = users[username]["image"]
        image = cv2.imread(image_path)

        if image is not None:
            image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

            # Display the image
            plt.figure(figsize=(6, 6))
            plt.title(f"Profile of {username}")
            plt.imshow(image_rgb)
            plt.axis('off')  # Hide axes
            plt.show()
        else:
            print("Error: Image file not found.")
    else:
        print("Incorrect password.")
else:
    print("User not found.")
```
